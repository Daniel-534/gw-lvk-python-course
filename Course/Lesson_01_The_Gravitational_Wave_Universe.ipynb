## 1. Introduction

On **14 September 2015**, at 09:50:45 UTC, the two LIGO detectors in Hanford (Washington) and Livingston (Louisiana) simultaneously recorded a transient signal that lasted less than half a second. That signal ŌĆö later named **GW150914** ŌĆö was the first direct detection of a gravitational wave (GW) in human history. It originated from the merger of two black holes each roughly 30 solar masses, located about 1.3 billion light-years from Earth. The announcement, made on 11 February 2016, marked the dawn of **gravitational-wave astronomy**.

This discovery was the culmination of a century of theoretical development ŌĆö beginning with Einstein's 1916 prediction ŌĆö and four decades of painstaking experimental work. It earned the 2017 Nobel Prize in Physics for **Rainer Weiss**, **Barry C. Barish**, and **Kip S. Thorne**.

### Why gravitational waves matter

Gravitational waves open a fundamentally new observational window onto the Universe. Unlike electromagnetic radiation, they:

- **Penetrate all matter without absorption.** A GW from the early Universe or from deep inside a collapsing star arrives at Earth essentially unimpeded.
- **Carry information about spacetime geometry itself**, not just the matter and fields embedded in it.
- **Are generated by bulk mass motion**, including objects (black holes) that emit no light at all.
- **Enable multi-messenger astronomy** when combined with electromagnetic and neutrino observations.

### Scope of this lesson

This first lesson is entirely theoretical. We will:

1. Develop the physics of gravitational waves from first principles (within the weak-field approximation).
2. Understand how km-scale laser interferometers detect the unimaginably small spacetime distortions they produce.
3. Survey the major observing runs and their scientific highlights.
4. Introduce the Gravitational-Wave Open Science Center (GWOSC), the public data repository that will be central to all subsequent lessons.

No programming is required here. All exercises at the end are analytic or conceptual.

## 2. Basic Physics of Gravitational Waves

### 2.1 General Relativity in a Nutshell

Einstein's **General Theory of Relativity** (GR), published in 1915, describes gravity not as a force but as the **curvature of four-dimensional spacetime** caused by the presence of mass and energy. The central equation is:

$$
G_{\mu\nu} \equiv R_{\mu\nu} - \frac{1}{2}g_{\mu\nu}R = \frac{8\pi G}{c^4}\, T_{\mu\nu}
$$

where:

| Symbol | Meaning |
|--------|---------|
| $g_{\mu\nu}$ | metric tensor (encodes distances in spacetime) |
| $R_{\mu\nu}$ | Ricci curvature tensor |
| $R$ | Ricci scalar |
| $G_{\mu\nu}$ | Einstein tensor (left-hand side = geometry) |
| $T_{\mu\nu}$ | stressŌĆōenergy tensor (right-hand side = matter/energy) |
| $G$ | Newton's gravitational constant |
| $c$ | speed of light |

The equation says, succinctly: **matter tells spacetime how to curve; curved spacetime tells matter how to move** (J. A. Wheeler).

Freely falling objects follow **geodesics** ŌĆö the generalisation of straight lines in curved spacetime. In the Newtonian limit, the geodesic equation reduces to Newton's second law with gravitational acceleration.

### 2.2 Linearised Gravity and the Wave Equation

The full Einstein equations are nonlinear and extremely hard to solve. For sources that are not extremely compact (or for the wave zone far from a compact source), we can write the metric as a small perturbation on flat (Minkowski) spacetime:

$$
g_{\mu\nu} = \eta_{\mu\nu} + h_{\mu\nu}, \qquad |h_{\mu\nu}| \ll 1
$$

where $\eta_{\mu\nu} = \mathrm{diag}(-1,+1,+1,+1)$ is the Minkowski metric and $h_{\mu\nu}$ is the perturbation.

Expanding the Einstein equations to first order in $h_{\mu\nu}$ and choosing the **Lorenz (harmonic) gauge** $\partial^\mu \bar{h}_{\mu\nu} = 0$ (where $\bar{h}_{\mu\nu} \equiv h_{\mu\nu} - \frac{1}{2}\eta_{\mu\nu}h$ is the trace-reversed perturbation), one obtains:

$$
\Box\, \bar{h}_{\mu\nu} = -\frac{16\pi G}{c^4}\, T_{\mu\nu}
$$

This is a **wave equation** driven by the stressŌĆōenergy tensor. In vacuum ($T_{\mu\nu} = 0$), the solution is a wave propagating at the speed of light:

$$
\Box\, \bar{h}_{\mu\nu} = 0 \quad \Longrightarrow \quad h_{\mu\nu} = A_{\mu\nu}\, e^{i k_\alpha x^\alpha}
$$

with $k^\alpha k_\alpha = 0$ (null wave vector), confirming that gravitational waves travel at $c$.

Further imposing **transverse-traceless (TT) gauge** removes all remaining gauge freedom. In TT gauge, the only non-zero components of $h_{\mu\nu}$ for a wave propagating in the $z$-direction are $h_{xx} = -h_{yy}$ and $h_{xy} = h_{yx}$, usually written:

$$
h^{\rm TT}_{\mu\nu} =
\begin{pmatrix}
0 & 0 & 0 & 0 \\
0 & h_+ & h_\times & 0 \\
0 & h_\times & -h_+ & 0 \\
0 & 0 & 0 & 0
\end{pmatrix}
\cos\!\left[\omega(t - z/c)\right]
$$

The two independent components $h_+$ and $h_\times$ are the **two polarisation states** of gravitational waves.

### 2.3 Polarisations: $h_+$ and $h_\times$

A gravitational wave stretches and squeezes spacetime **transverse** to its direction of propagation. The two polarisations act on a ring of test particles as follows:

**Plus polarisation ($h_+$):** squeezes in one transverse direction while stretching in the perpendicular direction (aligned with the coordinate axes).

**Cross polarisation ($h_\times$):** same pattern but **rotated by 45┬░**.

For a wave with frequency $f$ and amplitude $h_0$ propagating along $\hat{z}$:

$$
h_+(t) = h_0 \cos(2\pi f t), \qquad h_\times(t) = h_0 \sin(2\pi f t)
$$

(for circularly polarised emission as from a face-on binary).

The **strain** $h$ is the fractional change in proper distance between two freely-falling test masses separated by a distance $L$:

$$
h = \frac{\Delta L}{L}
$$

For GW150914, $h \approx 10^{-21}$ and $L = 4\,\mathrm{km}$, giving $\Delta L \approx 4 \times 10^{-18}\,\mathrm{m}$ ŌĆö about $\tfrac{1}{1000}$ the diameter of a proton.

**Spin of the graviton:** The two-polarisation structure is a consequence of the fact that the gravitational field is a spin-2 field. For a massless spin-$s$ field propagating in a given direction, there are exactly two helicity states: $\pm s$. For spin-2, these correspond to $h_+$ and $h_\times$.

### 2.4 The Quadrupole Formula

For a non-relativistic, weakly-gravitating source, the leading-order radiation is **quadrupolar** (monopole radiation is forbidden by mass conservation; dipole radiation is forbidden by momentum conservation). The radiated power is given by the **Einstein quadrupole formula**:

$$
P_{\rm GW} = \frac{G}{5c^5} \left\langle \dddot{Q}_{ij}\, \dddot{Q}^{ij} \right\rangle
$$

where $Q_{ij}$ is the **reduced quadrupole moment tensor**:

$$
Q_{ij} = \int \rho\left(x_i x_j - \frac{1}{3}\delta_{ij} r^2\right) d^3x
$$

and dots denote time derivatives. The three time derivatives ("jerk" of the mass distribution) mean that only **time-varying, non-spherically-symmetric** mass distributions radiate.

**Key consequences:**
- A spherically symmetric collapse (e.g., a perfectly symmetric supernova) emits **no** gravitational radiation.
- A uniformly rotating body with no asymmetry emits **no** gravitational radiation.
- Two masses orbiting each other in a non-circular orbit, or a wobbling/precessing neutron star, **do** radiate.

The corresponding strain amplitude at distance $r$ is:

$$
h \sim \frac{G}{c^4 r} \ddot{Q}
$$

Numerically, $G/c^4 \approx 8.26 \times 10^{-45}\,\mathrm{s^2/(kg\cdot m)}$, which explains why kilometre-scale detectors are needed even for astrophysical sources.

### 2.5 Characteristic Strain and Amplitude

For a **compact binary inspiral** (two point masses $m_1$, $m_2$ in a circular orbit), the gravitational wave frequency is twice the orbital frequency:

$$
f_{\rm GW} = 2 f_{\rm orb}
$$

The orbital frequency evolves according to:

$$
\dot{f}_{\rm orb} = \frac{96}{5} \pi^{8/3} \left(\frac{G\mathcal{M}}{c^3}\right)^{5/3} f_{\rm orb}^{11/3}
$$

where the **chirp mass** is:

$$
\mathcal{M} = \frac{(m_1 m_2)^{3/5}}{(m_1 + m_2)^{1/5}}
$$

The chirp mass is the **single most measurable parameter** of a compact binary: it determines both the rate at which the frequency sweeps upward (the "chirp") and the amplitude of the gravitational wave.

The peak strain from a binary merger at luminosity distance $d_L$ is approximately:

$$
h_{\rm peak} \approx \frac{4G}{c^2} \frac{\mathcal{M}}{d_L} \left(\frac{G\mathcal{M}\pi f_{\rm peak}}{c^3}\right)^{2/3}
$$

**Observation:** The strain $h$ decreases as $1/d_L$. Doubling detector sensitivity doubles the detectable volume (distance), hence increasing the detectable number of events by a factor of **8** in a volume-limited survey.

### 2.6 Sources of Gravitational Waves

Gravitational waves span an enormous frequency range and can be produced by many astrophysical processes.

| Frequency Band | Typical Frequency | Source | Detector |
|---|---|---|---|
| Ultra-low frequency | $10^{-18}$ŌĆō$10^{-15}$ Hz | Primordial GWs (inflation) | CMB B-modes (e.g., BICEP) |
| Very low frequency | $10^{-9}$ŌĆō$10^{-6}$ Hz | Supermassive BH binaries | Pulsar Timing Arrays (NANOGrav, EPTA, PPTA) |
| Low frequency | $10^{-4}$ŌĆō$10^{-1}$ Hz | Stellar-mass binaries (early inspiral), EMRIs | Space interferometers (LISA, TianQin, Taiji) |
| High frequency | $10$ŌĆō$10^3$ Hz | Compact binary mergers, NSs, supernovae | Ground-based interferometers (LIGO, Virgo, KAGRA) |

**Compact Binary Coalescences (CBCs):**
- **Binary Black Holes (BBH):** The most commonly detected type. Masses from ~5 to several hundred $M_\odot$.
- **Binary Neutron Stars (BNS):** Neutron star masses ~1.1ŌĆō2.0 $M_\odot$. Associated with short gamma-ray bursts and kilonovae (confirmed by GW170817).
- **Neutron Star ŌĆō Black Hole (NSBH):** First confirmed detections were NSBH events GW200105 and GW200115 in O3.

**Continuous Gravitational Waves (CW):** Nearly monochromatic signals from rapidly spinning neutron stars with a mass asymmetry (e.g., mountains on the crust, or unstable r-modes). None confirmed yet, but upper limits constrain neutron star physics.

**Burst Sources:** Short-duration, poorly-modelled signals from core-collapse supernovae, cosmic string cusps, and other exotic phenomena.

**Stochastic Background:** An incoherent superposition of unresolved sources from throughout cosmic history, potentially including a primordial background from the early Universe.

**Energy transport:** A compact binary merger releases an enormous amount of energy in gravitational waves. GW150914 radiated approximately $3\,M_\odot c^2 \approx 5.4 \times 10^{47}\,\mathrm{J}$ in less than a second ŌĆö more than the luminosity of all stars in the observable Universe combined, for that brief instant.

## 3. How Ground-Based Detectors Work

### 3.1 The Michelson Interferometer Principle

A **laser interferometer gravitational-wave detector** is, at its core, a very sophisticated **Michelson interferometer**. The principle is elegant:

1. A laser beam is split into two perpendicular arms by a **beam splitter**.
2. Each beam travels down an arm, reflects off a mirror at the far end, and returns.
3. The two beams recombine at the beam splitter. If the two arm lengths are equal, the beams cancel at the **dark port** (destructive interference) and all light returns to the laser (**bright port**).
4. A gravitational wave stretches one arm while squeezing the other. This introduces a path-length difference $\delta L = h L$ (where $L$ is the arm length), changing the interference pattern.
5. The resulting light at the dark port is detected by a photodiode. The signal is proportional to $h$.

The fundamental sensitivity limit of a simple Michelson with laser power $P$ at wavelength $\lambda$ is set by **photon shot noise**:

$$
h_{\rm shot} \sim \sqrt{\frac{\hbar c \lambda}{P L^2}}
$$

For LIGO-like parameters ($L = 4\,\mathrm{km}$, $\lambda = 1064\,\mathrm{nm}$, $P = 100\,\mathrm{W}$), this gives $h_{\rm shot} \sim 10^{-21}\,\mathrm{Hz}^{-1/2}$, which is still not enough. Several enhancements are therefore essential.

### 3.2 FabryŌĆōP├®rot Cavities and Power Recycling

**FabryŌĆōP├®rot (FP) Arm Cavities:**  
Each arm contains a pair of highly reflective mirrors that form an optical **resonant cavity**. Light bounces back and forth hundreds of times before leaking out, effectively multiplying the optical path length by the **finesse** $\mathcal{F}$ (typically ~450 for Advanced LIGO). This increases the effective arm length to $L_{\rm eff} = \mathcal{F} L / \pi \approx 1100\,\mathrm{km}$.

**Power Recycling Mirror (PRM):**  
The beam splitter normally operates at the dark fringe, sending most light back toward the laser. A **power recycling mirror** placed between the laser and the beam splitter forms a resonant cavity that recycles this light back into the interferometer. Advanced LIGO achieves ~280 kW of circulating power in each arm cavity from a ~200 W input laser.

The effective strain sensitivity improves as $\sqrt{P_{\rm circulating}}$, so the 1400├Ś power enhancement improves sensitivity by a factor of ~37.

### 3.3 Signal Recycling and Squeezing

**Signal Recycling Mirror (SRM):**  
A mirror placed at the dark port resonantly enhances the signal sidebands at frequencies of interest. In Advanced LIGO it operates in **broadband resonant sideband extraction (RSE)** mode, broadening the detector's frequency response.

**Quantum Squeezing:**  
Shot noise and radiation pressure noise arise from the quantum nature of light. In the standard quantum limit (SQL), there is a trade-off: squeezing the phase quadrature (reducing shot noise) amplifies amplitude noise (radiation pressure), and vice versa.  
Advanced LIGO and Virgo now inject **frequency-dependent squeezed light** using a filter cavity. This allows simultaneous reduction of shot noise at high frequencies and radiation pressure noise at low frequencies, improving sensitivity by up to 3 dB across the band.

### 3.4 Fundamental Noise Sources

The detector sensitivity is limited by a variety of noise sources, each dominant in different frequency bands:

**Seismic noise (< ~10 Hz):** Ground motion from microseism, human activity, and ocean waves couples into the detector through the suspension system. Below ~10 Hz the detector is essentially blind.

**Newtonian / Gravity gradient noise (~1ŌĆō30 Hz):** Fluctuating local gravitational fields (due to atmospheric and seismic density variations) directly couple to the test masses without going through the suspension, and cannot be shielded. This is a fundamental limit for ground-based detectors below ~10 Hz.

**Suspension thermal noise (~10ŌĆō100 Hz):** Thermal fluctuations in the fibres suspending the mirrors introduce position noise. Advanced LIGO uses **fused silica fibres** with very high mechanical Q to narrow the thermal noise peak.

**Mirror thermal noise (~30ŌĆō200 Hz):** Brownian motion of the mirror coating and substrate creates surface fluctuations. Coating thermal noise is currently the dominant noise source in the mid-band (~100 Hz). Reducing it is a major R&D goal for next-generation detectors.

**Quantum (shot) noise (> ~100 Hz):** Photon counting statistics at the photodiode. Improves with laser power and squeezing.

**Quantum (radiation pressure) noise (< ~50 Hz):** Photon momentum kicks on the mirrors from the high-power laser. Worsens with laser power.

These contributions combine to form the **amplitude spectral density (ASD)** noise curve, typically plotted as strain noise $\tilde{h}(f)$ in units of $\mathrm{Hz}^{-1/2}$. A signal is detectable when its power rises above the noise floor in a matched-filter sense.

### 3.5 Seismic Isolation and Suspension Systems

To protect the delicate mirrors from ground motion, every test mass is suspended at the bottom of a multi-stage isolation system:

**Advanced LIGO Quad Suspension:**  
A 4-stage pendulum system, with the mirror hanging from fused silica fibres at the final stage. Each stage provides passive isolation that rolls off as $f^{-2}$ above its resonant frequency. Four stages give $(f_0/f)^8$ isolation, reducing seismic noise by many orders of magnitude above ~10 Hz. Active feedback (using inertial sensors and actuators at the upper stages) controls low-frequency motion.

**Virgo Superattenuator:**  
An inverted pendulum at the top of a chain of mechanical filters provides active and passive isolation. The system is ~10 m tall and achieves similar performance to LIGO's quad suspension.

**KAGRA Cryogenic Suspension:**  
KAGRA cools its mirrors to ~20 K to reduce thermal noise. The mirrors are sapphire crystals suspended by sapphire fibres. Cryogenics introduce unique challenges (vibration from refrigerators, cryogenic mechanical loss) but offer a path to lower thermal noise.

### 3.6 LIGO

The **Laser Interferometer Gravitational-wave Observatory (LIGO)** consists of two identical detectors:

- **LIGO Hanford Observatory (LHO / H1):** Near Hanford, Washington State, USA
- **LIGO Livingston Observatory (LLO / L1):** Near Livingston, Louisiana, USA

**Separation:** ~3000 km (light travel time: ~10 ms)

**Key parameters (Advanced LIGO):**
- Arm length: 4 km
- Laser wavelength: 1064 nm (Nd:YAG)
- Input laser power: ~200 W
- Circulating arm power: ~280 kW
- Mirror mass: 40 kg (fused silica)
- Design sensitivity: $\sim 10^{-23}\,\mathrm{Hz}^{-1/2}$ at ~100 Hz

LIGO is operated by Caltech and MIT under a cooperative agreement with the NSF. The LIGO Scientific Collaboration (LSC) involves ~1000 scientists from ~100 institutions worldwide.

**History:** The initial LIGO (iLIGO) operated 2002ŌĆō2010, reaching design sensitivity but making no detections. Advanced LIGO (aLIGO), with a factor of ~10 improvement in strain sensitivity, began operations in 2015 and made the first detection immediately.

### 3.7 Virgo

**Virgo** is located near Cascina (Pisa), Italy, and is operated by the European Gravitational Observatory (EGO), a collaboration between France (CNRS) and Italy (INFN). The Virgo Collaboration also includes many other European and international institutions.

**Key parameters (Advanced Virgo+):**
- Arm length: 3 km
- Laser wavelength: 1064 nm
- Mirror mass: 42 kg (fused silica)
- **Superattenuator** seismic isolation
- Frequency-dependent squeezing (O4)

Virgo joined LIGO for the final weeks of O2, enabling the first three-detector observation (GW170814). Its role in the network is crucial for **sky localisation**: because Virgo is on a different continent, the time delay between LIGO and Virgo greatly constrains the sky position of sources.

### 3.8 KAGRA

**KAGRA** (KAmioka GRAvitational wave detector) is located underground in the Kamioka mine in Gifu Prefecture, Japan. It is operated by the Institute for Cosmic Ray Research (ICRR) at the University of Tokyo, and developed by ICRR, KEK, and NAOJ.

**Unique features:**
- **Underground location:** Reduces seismic noise and Newtonian noise by many orders of magnitude; also shields from temperature fluctuations and reduces weather-related noise.
- **Cryogenic mirrors:** Sapphire test masses cooled to ~22 K reduce thermal noise in the coating and mirror substrate.
- **Arm length:** 3 km

KAGRA first reached design sensitivity in 2020 and joined LIGOŌĆōVirgo for a brief observing period in late O3. As it matures, its geographic location (Japan vs. US and Europe) adds a longer East-West baseline and improved sky coverage for the global network.

### 3.9 The Global Network

Operating multiple detectors simultaneously is essential for several reasons:

1. **Noise rejection:** A genuine GW signal must appear in all detectors within the light-travel time between them. Coincident detection rejects local noise transients.
2. **Sky localisation:** The arrival-time differences between detectors triangulate the source position on the sky (analogous to GPS but using light travel time). With 2 detectors, the position is a ring on the sky; with 3+ detectors it becomes a patch of order $1$ŌĆō$100\,\mathrm{deg}^2$.
3. **Polarisation decomposition:** With 3+ non-coplanar detectors, both $h_+$ and $h_\times$ can be measured independently, testing GR and reconstructing the binary inclination angle.
4. **Duty cycle:** Different detectors have different downtime for maintenance; coincident requirements mean that total network observation time is less than individual uptime.

**LIGO-India** (LIGO-Aundha, Maharashtra) is approved and under construction, expected to join the network ~2030. Its addition to the LIGOŌĆōVirgoŌĆōKAGRA (LVK) network will dramatically improve sky localisation (median area ~2ŌĆō5 deg$^2$) and enable near-complete polarisation measurement.

**Next-generation detectors:**
- **Einstein Telescope (ET):** A proposed underground, triangular, cryogenic interferometer in Europe with ~10 km arms, targeting a factor of ~10 improvement over Advanced LIGO.
- **Cosmic Explorer (CE):** Proposed US detector with 20 km (or 40 km) arm length, surface-based, expected to be sensitive to BNS mergers throughout the Universe.

## 4. Observing Runs: O1 through O4

Gravitational-wave detectors do not operate continuously. Between observing runs ("O" periods), the instruments undergo extensive hardware upgrades and commissioning. The naming convention is **O1, O2, O3a, O3b, O4a, O4b**, etc.

### 4.1 O1 (2015ŌĆō2016)

**Dates:** 12 September 2015 ŌĆō 19 January 2016 (~4 months)  
**Active detectors:** LIGO H1, LIGO L1  
**Network strain sensitivity (BNS range):** ~75 Mpc

**Key events:**

| Event | Type | Date | Total Mass ($M_\odot$) | Distance (Mpc) | Notes |
|---|---|---|---|---|---|
| **GW150914** | BBH | Sep 14, 2015 | ~71 | ~410 | **First GW detection**, Nobel Prize |
| **GW151226** | BBH | Dec 26, 2015 | ~21 | ~440 | "Boxing Day" event; smaller masses, ~55 cycles visible |
| LVT151012 | BBH | Oct 12, 2015 | ~35 | ~1000 | Marginal candidate (later confirmed as GW151012 in GWTC-1 re-analysis) |

**Scientific highlights:**
- First direct detection of gravitational waves (GW150914, SNR ~24).
- First observation of a stellar-mass black hole binary.
- Confirmation that binary black holes merge within a Hubble time.
- Tests of GR in the strong-field, high-velocity regime (no deviations found).
- Upper limit on graviton mass: $m_g < 1.2 \times 10^{-22}\,\mathrm{eV}/c^2$.

### 4.2 O2 (2016ŌĆō2017)

**Dates:** 30 November 2016 ŌĆō 25 August 2017 (~9 months)  
**Active detectors:** LIGO H1, LIGO L1; Virgo joined 1 August 2017  
**Network BNS range:** ~80ŌĆō100 Mpc (LIGO); ~25ŌĆō30 Mpc (Virgo at start)

**Key events:**

| Event | Type | Date | Notes |
|---|---|---|---|
| **GW170104** | BBH | Jan 4, 2017 | Heaviest BBH at the time; hints of spin misalignment |
| **GW170608** | BBH | Jun 8, 2017 | Lightest BBH in O2; detected in H1 and L1 |
| **GW170814** | BBH | Aug 14, 2017 | **First three-detector detection** (H1+L1+V1); first GW polarisation test |
| **GW170817** | BNS | Aug 17, 2017 | **First multi-messenger GW event** (+ GRB 170817A + kilonova AT2017gfo) |

**GW170817 ŌĆö A Landmark Event:**  
GW170817 was detected 1.7 seconds before the short gamma-ray burst GRB 170817A seen by Fermi and INTEGRAL. This delay constrains the difference between the speed of gravity and light to $|v_{GW} - c|/c < 5 \times 10^{-16}$, ruling out many modified gravity theories. The optical counterpart (kilonova AT2017gfo) was observed in NGC 4993 at ~40 Mpc, representing the first observation of heavy element synthesis (r-process) in a neutron star merger. GW170817 also provided the first GW-based measurement of the Hubble constant: $H_0 = 70^{+12}_{-8}\,\mathrm{km\,s^{-1}\,Mpc^{-1}}$.

**Total O2 catalogue (GWTC-1):** 11 GW events (10 BBH + 1 BNS).

### 4.3 O3 (2019ŌĆō2020)

**Dates:**  
- O3a: 1 April 2019 ŌĆō 1 October 2019  
- O3b: 1 November 2019 ŌĆō 27 March 2020 (ended 2 weeks early due to COVID-19)  
**Active detectors:** LIGO H1, LIGO L1, Virgo V1; KAGRA joined for ~2 weeks at end of O3b  
**Network BNS range:** ~120 Mpc (LIGO); ~50ŌĆō65 Mpc (Virgo)

O3 saw a dramatic increase in event rate (~1 event per week) due to improved sensitivity. A real-time public alert system was implemented for the first time, allowing electromagnetic follow-up of GW candidates within minutes.

**Selected notable events:**

| Event | Type | Notes |
|---|---|---|
| GW190412 | BBH | **Asymmetric mass ratio** (4:1); first detection of higher-order multipoles |
| GW190425 | BNS | Total mass ~3.4 $M_\odot$; much heavier than known Galactic BNS systems |
| GW190521 | BBH | Total mass ~150 $M_\odot$; remnant in **intermediate-mass BH** range (>100 $M_\odot$); possible electromagnetic counterpart |
| GW190814 | NSBH? | Mass ratio 9:1; secondary ~2.6 $M_\odot$ (either heaviest NS or lightest BH) |
| GW200105, GW200115 | NSBH | **First confirmed NSBH detections** |

**Catalogues from O3:**
- GWTC-2 (O1+O2+O3a): 47 events
- GWTC-2.1 (deep reanalysis of O3a): 55 events
- GWTC-3 (O1+O2+O3): 90 confident events

**Scientific highlights:**
- Confirmation of NSBH binaries as a new class of GW sources.
- Evidence for a gap in the black hole mass spectrum (~50ŌĆō130 $M_\odot$, the pair-instability mass gap).
- First detections of higher-order multipoles ($\ell=3,4$), allowing new GR tests.
- Population analyses revealing the BBH mass distribution, spin distribution, and merger rate as a function of redshift.

### 4.4 O4 (2023ŌĆōpresent)

**Dates:**  
- O4a: 24 May 2023 ŌĆō 16 January 2024  
- O4b: March 2024 ŌĆō ongoing (as of this writing)  
**Active detectors:** LIGO H1, LIGO L1, Virgo V1, KAGRA  
**Network BNS range:** ~160ŌĆō190 Mpc (LIGO A+); ~40ŌĆō50 Mpc (Virgo)

**Key upgrades for O4:**
- **LIGO A+:** Frequency-dependent squeezing (filter cavity), improved input optics, better stray light control ŌåÆ ~20ŌĆō40% sensitivity improvement over O3 LIGO.
- **Advanced Virgo+:** Phase 1 upgrades including better signal recycling mirrors and frequency-dependent squeezing in O4b.
- Fully automated public alert system sending low-latency candidates to the Gamma-ray Coordinates Network (GCN) within minutes.

By early 2024, O4 had already surpassed the total number of events from all previous runs. The event rate in O4 is expected to be ~10ŌĆō20├Ś higher than O1. Early results include multiple BNS candidates and new NSBH events.

### 4.5 Sensitivity Evolution

The following table summarises the BNS inspiral range (the distance at which an optimally-oriented, face-on BNS merger produces a network signal-to-noise ratio of 8) across observing runs:

| Run | LIGO BNS range | Virgo BNS range | Total confident events |
|---|---|---|---|
| O1 | ~75 Mpc | ŌĆö | 3 |
| O2 | ~80ŌĆō100 Mpc | ~25ŌĆō30 Mpc | 11 |
| O3 | ~110ŌĆō130 Mpc | ~45ŌĆō65 Mpc | ~79 |
| O4 | ~160ŌĆō190 Mpc | ~40ŌĆō50 Mpc | >100 (ongoing) |

Since the observable volume scales as $d^3$, the factor of ~2.5 improvement from O1 to O4 in LIGO's range corresponds to a factor of ~15 increase in accessible volume.

**Road to design sensitivity:** The Advanced LIGO design sensitivity specifies a BNS range of ~200 Mpc. Achieving this requires reducing coating thermal noise (through new coating materials such as amorphous silicon), further squeezing enhancements, and continued commissioning.

## 5. The GWOSC Ecosystem

### 5.1 What Is GWOSC?

The **Gravitational-Wave Open Science Center (GWOSC)**, hosted at [gwosc.org](https://gwosc.org), is the public data repository of the LIGOŌĆōVirgoŌĆōKAGRA collaboration. It was launched in 2015 (as LOSC ŌĆö LIGO Open Science Center) and expanded to include Virgo and KAGRA data.

**Mission:** Make GW data freely accessible to the global scientific community and the public, in the spirit of open science.

**What GWOSC provides:**
1. **Calibrated strain data** from all LVK detectors, in all observing runs.
2. **Event catalogues** with parameter estimates and posterior samples.
3. **Data quality flags and segment lists** indicating when each detector was in observing mode.
4. **Tutorials, documentation, and datasets** for educational purposes.
5. **Software tools** for accessing and analysing the data.

**Data policy:** All GW strain data are released publicly on a schedule defined by the LVK. For O3, data were released ~18 months after collection. For O4, the latency has been reduced. Some data (especially around gravitational-wave events) are released more quickly for follow-up analyses.

### 5.2 Event Catalogues

GWOSC hosts all published GW event catalogues:

| Catalogue | Covering | Events | Reference |
|---|---|---|---|
| GWTC-1 | O1 + O2 | 11 | Abbott et al. (2019), PRX 9, 031040 |
| GWTC-2 | O1 + O2 + O3a | 47 | Abbott et al. (2021), PRX 11, 021053 |
| GWTC-2.1 | O1 + O2 + O3a (deep) | 55 | Abbott et al. (2024), PRD 109, 022001 |
| GWTC-3 | O1 + O2 + O3 | 90 | Abbott et al. (2023), PRX 13, 041039 |

Each event in the catalogue has:
- A **GW name** in the format `GW` + date (YYMMDD) + optional letter suffix (e.g., GW150914, GW191109A).
- **Parameter estimates:** component masses, spins, distance, sky location, etc., provided as posterior probability distributions (sampled using Bayesian inference).
- **Strain data** around the event time (typically ┬▒4096 s).
- A **classification** of the source type (BBH, BNS, NSBH, unmodelled burst).
- Probabilities $p_\mathrm{astro}$ (probability of astrophysical origin) and FAR (false alarm rate in events per year).

### 5.3 Open Data: Strain Files and Metadata

**Strain data format:**  
Data are provided as time series of gravitational wave strain $h(t)$ sampled at 4096 or 16384 Hz, stored in two formats:

- **HDF5 (.hdf5 / .h5):** Hierarchical Data Format 5. Self-describing, portable, supports compression. Preferred for programmatic access.
- **GWF (.gwf):** Frame format, native to the GW community. Requires the `lalframe` or `framel` libraries.

**Segment lists:**  
Each detector has **state flags** indicating its observational mode:
- `CBC_CAT1`: detector in observing mode, data suitable for CBC searches.
- `CBC_CAT2`: additional data quality cuts applied.
- `CBC_CAT3`: further cuts (hardware injections excluded, etc.).
Segment lists (lists of GPS time intervals satisfying each flag) are available on GWOSC and can be queried programmatically.

**GPS Time:**  
All LVK data use **GPS time**: number of seconds since 6 January 1980 00:00:00 UTC (no leap seconds). GW150914 occurred at GPS time 1126259462.4. The conversion between GPS and UTC is well-defined but requires accounting for leap seconds.

**Data channels:**  
The main science channel is the **gravitational-wave channel** (e.g., `H1:GDS-CALIB_STRAIN` for LIGO Hanford). Hundreds of **auxiliary channels** monitor environmental and instrumental effects (seismometers, magnetometers, microphones, etc.) and are used for noise characterisation and glitch studies.
